```python
import matplotlib.pyplot as plt
%matplotlib inline
filename_in =  '/Volumes/alexandr/palmer/data_store/IMS/centroids/20150730_ANB_spheroid_control_65x65_15um_centroids_IMS.hdf5'
# Parse data
from pyIMS.hdf5.inMemoryIMS_hdf5 import inMemoryIMS_hdf5
from pyIMS.image_measures import level_sets_measure
IMS_dataset=inMemoryIMS_hdf5(filename_in)
```

```python
# Get images from a list of m/zs
import numpy as np
ppm=1. #mass tolerance in parts per million
mz_list=np.asarray([798.54096675,799.54432915,800.54419457,])
ion_datacube = IMS_dataset.get_ion_image(mz_list,ppm)
#Show images
for ii,mz in enumerate(mz_list):
    plt.figure()
    im = ion_datacube.xic_to_image(ii)
    plt.imshow(im)
    plt.title(mz)
    plt.show()
```
